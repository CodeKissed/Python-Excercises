### PRACTICE DATA WRANGLING

This notebook shows some common data wrangling techniques. The dataset used contains the prices and other attributes of almost 54,000 diamonds. The dataset can be accesed through this link: https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv.

Below is a  Pandas practice set found in https://www.w3resource.com/python-exercises/pandas/practice-set1/index.php which I find very useful for beginners to be familiar with the most common functions and methods in processing data. Most Python online courses do not usually cover all techniques shared in this practice data set.


```python
import pandas as pd
```

```python
diamonds = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv')
diamonds
```

```python
# Modify the default columns values and print the first 6 rows.

df = diamonds.loc[:,["carat", "cut", "x", "y", "z"]]
df.columns = df.columns.str.capitalize()
df.head(6)
```

```python
# Select a series from diamonds DataFrame.
diamonds["carat"]
```

```python
# Add a new column labeled 'Quality -color'
diamonds["quality - color"] = diamonds.cut + ' , ' + diamonds.color
diamonds.head()
```

```python
# Find the number of rows and columns and data type of each column of diamonds Dataframe.
diamonds.shape
```

```python
diamonds.dtypes
```

```python
# Summarize only 'object' columns of the diamonds Dataframe.
diamonds.describe(include =['object'])
```

```python
# Rename two of the columns of the df Dataframe
df.rename(columns={"Cut": "Diamond_Cut", "Carat": "Diamond_Carat"})
```

```python
# Rename all the columns of the df Dataframe.
# Option 1

cols = ["Diamond_Carat", "Diamond_Cut", "Diamond_X", "Diamond_Y", "Diamond_Z"]
df.set_axis([cols], axis=1)
```

```python
# Option 2
cols = ["Diamond_Carat", "Diamond_Cut", "Diamond_X", "Diamond_Y", "Diamond_Z"]
df.columns = cols
df
```

```python
#Remove the second column of the df Dataframe
df.drop(["Diamond_Cut"], axis =1)
```

```python
#Remove the multiple columns of the df Dataframe
df.drop(["Diamond_Cut", "Diamond_Carat"], axis =1)
```

```python
# Sort the 'cut' Series in ascending order (returns a Series) of df Dataframe
df.Diamond_Cut.sort_values(ascending =True)
```

```python
# Sort the 'cut' Series in descending order (returns a Series) of df Dataframe
df.Diamond_Cut.sort_values(ascending =False)
```

```python
# Sort the entire diamonds DataFrame by the 'carat' Series in ascending order.
df.sort_values("Diamond_Carat")
```

```python
# Sort the entire diamonds DataFrame by the 'carat' Series in descending order.
df.sort_values("Diamond_Carat", ascending = False)
```

```python
# Filter the DataFrame rows to only show carat weight at least 0.3.

df.query("Diamond_Carat >= 0.3")
```

```python
# Find the details of the diamonds where length>5, width>5 and depth>5

diamonds.query("x > 5 & y > 5 & z > 5 ")
```

```python
# Find the diamonds that are either Premium or Ideal

diamonds.query("cut == 'Premium' or cut == 'Ideal' ")
```

```python
# Find the diamonds that are with a Fair or Good or Premium
result = diamonds.query("cut == 'Fair' | cut == 'Good' | cut == 'Premium' ")

result.sort_values("cut", ascending=False)
```
